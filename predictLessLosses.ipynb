```python
import pandas as pd
from sklearn.metrics import accuracy_score
```

# Predict team with less losses
The loss column is how many losses each team has going into the game

```python
df = pd.read_csv('nbaDatabasePreProcessed2004-2024.csv')
df
```

```python
df['predicted_winner'] = df.apply(lambda row: row['homeTeam'] if row['homeTeam_losses'] <= row['awayTeam_losses'] else row['awayTeam'], axis=1)
df
```

```python
print(accuracy_score(df['winner'], df['predicted_winner']))
```
